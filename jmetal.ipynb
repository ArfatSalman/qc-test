```python
import os
from json import loads

files = os.listdir('data/jmetal-search-data')
data = [ loads( open('data/jmetal-search-data/' + filename).read())  for filename in files ]
```

```python
class Solution:
    def __init__(self, variables,objectives,constraints):
        self.objectives = objectives

jmetal_stats = [ eval(el['jmetal']['result']).objectives[0] * -1 for el in data ] 
existing_stats = [  ]
for el in data:
    stored  = el['stored_perm_info']
    for k in stored:
        existing_stats.append(stored[k]['divergence_from_cirq']['statistic'])

print(len(jmetal_stats))
print(len(existing_stats))

import matplotlib.pylab as plt

plt.plot(jmetal_stats, label="from jmetal")
plt.plot(existing_stats, label="existing")
plt.xticks(range(len(jmetal_stats)))
plt.legend(loc="upper right")
plt.show()
```
