```python
from json import loads
import os
import pandas as pd
import matplotlib.pylab as plt

v53 = "data/qmt_v53/programs/metadata/"
v52 = "data/qmt_v52/programs/metadata/"
cirq_circuits_folder = "data/qmt-cirq/cirq-src/"
exec_metadata_path = "data/qmt-cirq/exec-metadata/"
```

```python
files_v52 = os.listdir(v52)
files_v53 = os.listdir(v53)

def get_data(args):
    qubits = {}
    gates = {}
    for files, folder in args:
        for file in files:
            fullpath = folder +  file
            content = loads(open(fullpath).read())
            source = content['source']
            
            selected_gates = source['selected_gate_set']
            
            if source['n_qubits'] in qubits:
                qubits[source['n_qubits']].append(source['program_id'])
            else:
                qubits[source['n_qubits']] = [ source['program_id'] ] 

            if source['n_ops'] in gates:
                gates[source['n_ops']].append(source['program_id'])
            else:
                gates[source['n_ops']] = [ source['program_id'] ] 

    return qubits, gates
        
qubits, gates = get_data([ (files_v52, v52), (files_v53, v53) ])
```

```python
qbs = {}
for k in sorted(qubits):
    qbs[len(qubits[k])] = k
qbs
```

```python

lists = sorted(qbs.items()) # sorted by key, return a list of tuples

x, y = zip(*lists) # unpack a list of pairs into two tuples

plt.bar(y, x)
plt.title('qubit distribution')
plt.xlabel("no of qubits")
plt.ylabel("count")
plt.show()
```

```python
gts = {}
for k in sorted(gates):
    gts[len(gates[k])] = k
gts
```

```python
lists = sorted(gts.items()) # sorted by key, return a list of tuples

x, y = zip(*lists) # unpack a list of pairs into two tuples

plt.bar(y, x)
plt.xlabel("no of operations")
plt.ylabel("count")
plt.show()
```

```python
qubits_for_plot = []
gates_for_plot = []
for program_id in total_set:
    for k, v in qubits.items():
        if program_id in v:
            qubits_for_plot.append(k)
    for k, v in gates.items():
        if program_id in v:
            gates_for_plot.append(k)

print(len(qubits_for_plot) == 100)
print(len(gates_for_plot) == 100)
```

```python
x_axis = list(range(0, 100))

plt.plot(x_axis, qubits_for_plot)
plt.plot(x_axis, gates_for_plot)
plt.show()
```
