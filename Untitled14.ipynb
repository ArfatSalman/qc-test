```python
from lib.utils_db import get_database_connection
```

```python
con = get_database_connection({ "experiment_folder": 'data/qmt_v53' }, "qfl.db")
```

```python
import pandas as pd
```

```python
df = pd.read_sql("SELECT * FROM QFLDATA", con)
```

```python
pval_col = f"divergence.ks.p-value"
```

```python
df_sorted_pvals = df.sort_values(by=[pval_col])
```

```python
df_sorted_pvals
```

```python
k = len(df_sorted_pvals)
```

```python
a = next(df_sorted_pvals.iterrows())
```

```python
i_star = None
method = 'holm'
alpha_level = 0.05
how_many_loops = 10
j = 0;
for i, (idx, row) in enumerate(df_sorted_pvals.iterrows()):
        ordinal_i = i + 1
        P_i = row[pval_col]
        if method == 'holm':
            threshold = alpha_level / (k - ordinal_i + 1)
        elif method == 'bonferroni':
            threshold = alpha_level / (k)
        elif method == 'bh':
            threshold = (alpha_level / (k)) * ordinal_i
        print(f"P_i={P_i}    ordinal_i={ordinal_i}, k={k}, threshold={threshold}")
        if P_i > threshold:
            i_star = i
            print(f"i*: {i_star}")
            break
        
        if j > how_many_loops:
            break
        else:
            j += 1
```

```python
df_divergent = df_sorted_pvals.iloc[:i_star]
```
